### Load names_match_improved

```python
matches = pd.read_csv('name_match_improved.csv')
print(matches.shape)
matches.head(10)
```

```python
matches[(matches['score'] <= 8) & (matches['score'] >= 7)].sort_values(by='score')
```

From above, we can see that some where around 7.5 would be a good threshold to set so that the names are accurately matched

```python
threshold = 7.5 # for matching score 
matches = matches[matches['score'] >= threshold]
matches['littlesis'] = matches['littlesis'].str.lower().str.strip()

print(matches.shape)
matches.head(10)
```

### Merge together with relationships on name, result should be one(person) to many (people he's connected with)

```python
merged_entity = pd.merge(
    matches,
    df_relationships,
    right_on='entity1_name_norm',
    left_on='littlesis',
    how='inner'
)
print(merged_entity.shape)
merged_entity.head(10)
```
